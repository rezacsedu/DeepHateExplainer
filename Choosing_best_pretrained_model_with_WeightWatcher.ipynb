```python
## WeightWatcher helps you choose the best pretrained model for your needs.

## You can use WeightWatcher to compare several pretrained models and choose the one with the lowest Log Norm.
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = 8,8

from scipy.ndimage.interpolation import zoom
import numpy as np
import os
from PIL import Image
import weightwatcher as ww
```

```python
import torch
import torch.nn as nn
import transformers
from transformers import AutoModel, BertModel
import numpy as np

class BERTBengaliTwo(nn.Module):
    def __init__(self):
        super(BERTBengaliTwo, self).__init__()
        self.bert = BertModel.from_pretrained("sagorsarker/bangla-bert-base",output_hidden_states=True)
        self.drop_out = nn.Dropout(0.1) 
        self.l0 =  nn.Linear(768 * 2, 5)
        torch.nn.init.normal_(self.l0.weight, std=0.02)

    def forward(self, ids, attention_mask, token_type_ids):
        _, _, out = self.bert(
            ids, 
            attention_mask=attention_mask, 
            token_type_ids=token_type_ids
        )
        out = torch.cat((out[-1], out[-2]), dim=-1)
        #out = self.drop_out(out)
        out = out[:,0,:]
        logits = self.l0(out)
        return logits

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
bertPoolModel = BERTBengaliTwo()
bertPoolModel.to(device)
bertPoolModel.load_state_dict(torch.load("bangla-bert-pool-base-lr-3e-5.bin", map_location=torch.device('cpu')))
```

```python
bertPoolModelwatcher = ww.WeightWatcher(model=bertPoolModel)
bertPoolModelDetails = bertPoolModelwatcher.analyze()
bertPoolModelSummary = bertPoolModelwatcher.get_summary(bertPoolModelDetails)
```
